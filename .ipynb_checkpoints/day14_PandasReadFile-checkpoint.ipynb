```python
#MAX AND MIN
df['GAJI'].max()
df['GAJI'].min()
df['GAJI'].mean()
df['GAJI'].sum()
df['GAJI'].mode()
df['GAJI'].median()
```

```python
#FILLNA 
df = pd.read_csv(
    'datakosong.csv',
    na_values = ['-','n.a']
)
df = df.fillna({
    'nama':'Anonim',
    'usia':24,
    'gaji':5000000
})
df
```

```python
#FORWARD FILL DAN BACKWARD FILL
df = pd.read_csv(
    'datakosong.csv',
    na_values = ['-','n.a']
)
# df = df.fillna(method = 'ffill', axis=1)
df = df.fillna(method = 'bfill', axis=1)
df
```

```python
df = pd.read_csv(
    'datakosong.csv',
    na_values = ['-','n.a']
)

df = df.fillna(0)
df['gaji']=df['gaji'].astype('int64')
df
```

```python
df = df.interpolate()
df
```

```python
df = df.interpolate().fillna({
    'nama': 'Anonim'
})
df
```
