```python
#GROUP BY
df = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Putu', 'Fisika':95, 'Biologi': 80},
    {'nama':'Putu', 'Fisika':100, 'Biologi': 85},
    {'nama':'Kadek', 'Fisika':80, 'Biologi': 90},
    {'nama':'Kadek', 'Fisika':90, 'Biologi': 95},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 100},
])
df
```

```python
#CONCAT = untuk menggabungkan dua dataframe

#1. DF dengan properti yang sama 

df1 = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 80},
    {'nama':'Komang', 'Fisika':100, 'Biologi': 85},
])
df2 = pd.DataFrame([
    {'nama':'Ana', 'Fisika':80, 'Biologi': 90},
    {'nama':'Elsa', 'Fisika':90, 'Biologi': 95},
    {'nama':'Kristoff', 'Fisika':95, 'Biologi': 100},
])
dfX = pd.concat([df1,df2], ignore_index=True)
dfX
```

```python
dfX.loc[0]
```

```python
#2.DF DENGAN PROPERTI YANG BERBEDA
df1 = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 80},
    {'nama':'Komang', 'Fisika':100, 'Biologi': 85},
])
df2 = pd.DataFrame([
    {'nama':'Ana', 'Math':80, 'Biologi': 90},
    {'nama':'Elsa', 'Math':90, 'Biologi': 95},
    {'nama':'Kristoff', 'Math':95, 'Biologi': 100},
])
dfX = pd.concat([df1,df2], ignore_index=True, sort=False)
dfX
```

```python
#3.DF DENGAN ORANG YANG SAMA 
df1 = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 80},
    {'nama':'Komang', 'Fisika':100, 'Biologi': 85},
])
df2 = pd.DataFrame([
    {'nama':'Putu', 'Math':80, 'Kimia': 90},
    {'nama':'Kadek', 'Math':90, 'Kimia': 95},
    {'nama':'Komang', 'Math':95, 'Kimia': 100},
])
dfX = pd.concat([df1,df2], axis=1,ignore_index=True, sort=False)
dfX

#axis=1 supaya menunjuk pada orang yang sama. urutan nama df1 dan df2 harus sama. 
```

```python
#4.DENGAN ORANG YANG SAMA TAPI INDEX DF1 DAN DF2 TIDAK SAMA 
df1 = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 80},
    {'nama':'Komang', 'Fisika':100, 'Biologi': 85},
],index= [0,1,2])
df2 = pd.DataFrame([
    {'nama':'Kadek', 'Math':80, 'Kimia': 90},
    {'nama':'Putu', 'Math':90, 'Kimia': 95},
    {'nama':'Komang', 'Math':95, 'Kimia': 100},
],index=[1,0,2])
dfX = pd.concat([df1,df2], axis=1,ignore_index=True, sort=False)
dfX
```

```python
#MERGE =JOIN (yang punya hub saja yang dikeluarkan)
df1 = pd.DataFrame([
    {'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'nama':'Kadek', 'Fisika':95, 'Biologi': 80},
    {'nama':'Komang', 'Fisika':100, 'Biologi': 85},
])
df2 = pd.DataFrame([
    {'nama':'Kadek', 'Math':80, 'Kimia': 90},
    {'nama':'Putu', 'Math':90, 'Kimia': 95},
    {'nama':'Komang', 'Math':95, 'Kimia': 100},
])
dfX = pd.merge(df1,df2,on='nama')
dfX

#on = nama adalah sebagai connector. defaultnya adalah inner join. 
#how = 'outer' => semua elemen dalam data frame dikeluarkan
```

```python
#PIVOT
df1 = pd.DataFrame([
    {'Test':1,'nama':'Putu', 'Fisika':90, 'Biologi': 78},
    {'Test':2,'nama':'Putu', 'Fisika':95, 'Biologi': 80},
    {'Test':3,'nama':'Putu', 'Fisika':100, 'Biologi': 85},
    {'Test':1,'nama':'Kadek', 'Fisika':80, 'Biologi': 90},
    {'Test':2,'nama':'Kadek', 'Fisika':90, 'Biologi': 95},
    {'Test':3,'nama':'Kadek', 'Fisika':95, 'Biologi': 100},
])
df1
```

```python
df1 =df1.pivot(index='Test',columns='nama')
df1.columns
```

```python
#CARA MANUAL BUAT DF DIATAS

df = pd.DataFrame({
    'mapel':['biologi','biologi','biologi','fisika','fisika','fisika'],
    'test':[1,2,3,1,2,3],
    'Putu':[90,95,100,78,80,85],
    'Kadek':[80,90,95,90,95,100]
    
})
df = df.set_index(['test','mapel']).unstack()
df = df.swaplevel(0,1,1).sort_index(1)
df

#swaplevel(0,1,1) level 0 pindah ke 1 axis=1 
#unstack = bikin dataframe jadi single ?? 

# MultiIndex([('biologi', 'Kadek'),
#             ('biologi',  'Putu'),
#             ( 'fisika', 'Kadek'),
#             ( 'fisika',  'Putu')],
#            names=['mapel', None])
```
